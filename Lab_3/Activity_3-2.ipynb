1. A square of asterisks.

```python
for i in range(5):
    print("*"*5)
```

2. A right-angled triangle of asterisks.

```python
for i in range(1, 6):
    print("*"*i)
```

3.1 An inverted right-angled triangle.

```python
n = 5

for i in range(1, n+1):
    print(" " * (n-i), end="")
    print("*" * i)
```

3.2  a pyramid.

```python
n = 5

for i in range(1, n+1):
    print(" " * (n - i), end="")
    print("* " * i)
```
