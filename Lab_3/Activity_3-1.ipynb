Buggy Snippet A: Infinite Loop
เป็น infinite loop เพราะไม่มีการอัพเดทค่า counter
แก้ไขโดยการทำการเพิ่มค่า counter 1 ค่าทุกครั้งที่รัน

Buggy Snippet B: Incorrect Range
ฟังค์ชั่น range(a, b) จะทำการสร้างลิสต์ของตัวเลขที่เริ่มที่ a (เริ่มที่ 0 ในกรณีที่ใส่พาราเมเตอร์ตัวเดียว) และจบที่ b-1 ดังนั้นถ้าพาราเมเตอร์ตัวที่ 2 คือเลข 5 จะจบที่ 4 เช่น range(1, 5) = [1, 2, 3, 4]
จึงต้องใส่พาราเมเตอร์ตัวที่ 2 เป็น 6 หากต้องการหยุดที่ 5

Buggy Snippet C: Misplaced Break
ต้องการให้แสดงผล 'h', 'e', 'l' แล้วจึงเบรค แต่คำสั่งเจอ 'l' แล้วเบรคก่อนที่จะแสดงผล พอรันจึงได้
'h'
'e'
Found 'l', stopping!
แก้ไขโดยทำการแสดงผลก่อนค่อยตรวจสอบหา 'l'

```python
# Buggy Snippet A: Infinite Loop
counter = 0
while counter < 5:
    print(counter)
    counter += 1
```

```python
# Buggy Snippet B: Incorrect Range
for i in range(1, 6): # Goal: print 1, 2, 3, 4, 5
    print(i)
```

```python
# Buggy Snippet C: Misplaced Break
for char in "hello":
    print(char) # This will print 'h', 'e', 'l' then break
    if char == 'l':
        print("Found 'l', stopping!")
        break
```
